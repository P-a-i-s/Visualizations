```python
fig = plt.figure() #create a figure object (plot of land)
ax = plt.axes() #create the axes
x = np.linspace(0,10,1000) #generating data points on the graph (content)
ax.plot(x,np.sin(x), color = 'black') #plot x1y1 plot curve which is sin (x) here (create waves shape)
ax.plot(x,np.cos(x), color = 'brown') #plot x2y2 curve which is cos(x) 
```

### Line Colors and Styles

```python
plt.plot(x,np.sin(x),color = 'blue')
```

```python
plt.plot(x,np.cos(x),color='#91487B')
```

```python
plt.plot(x,np.cos(x),color='#91487B', linestyle = '-.')
```

```python
plt.plot(x,np.sin(x), '-.r')
```

### SETTING UP THE AXES

```python
plt.plot(x,np.sin(x))
```

```python
## Setting Limits
plt.plot(x,np.sin(x))
plt.xlim(0,10)
plt.ylim(0,1)
```

```python
plt.plot(x,np.sin(x),label='Sin(x)')
plt.axis('equal')
plt.title('My Graphs')
plt.xlabel('x')
plt.ylabel('Sin(x)')
```

### Legend

```python
plt.plot(x,np.sin(x),label = 'sin(x)')
plt.plot(x,np.cos(x),label = 'cos(x)')
plt.axis('equal')
#plotting the legend
plt.legend()
```
