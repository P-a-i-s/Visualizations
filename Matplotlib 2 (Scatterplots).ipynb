```python
rng = np.random.RandomState(10)
x=rng.randn(100)
y=rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100) #1000 cause else the representation will be too small
```

### NOTES:
    ALPHA = Opacity/transperency of the graph 
    CMAP = color coding option (magma, inferno, plasma etc)
    COLORS = Choose from list of colors available (red, green, blue etc)

```python
x = [1,2,3,4]
y = [5,6,7,8]
colors = [2.2,5,2.5,0]

plt.scatter(x,y, c=colors, s=sizes, alpha=0.9, cmap = 'plasma')
plt.colorbar()
```
