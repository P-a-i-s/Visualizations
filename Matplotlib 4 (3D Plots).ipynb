```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,50)
y = np.linspace(0,5,40)

#MESHGRID
#makes 2 dimensional matrix using two 1-D arrays
X,Y = np.meshgrid(x,y)
```

```python
X.shape
```

```python
Y.shape
```

```python
X
```

### SHOWING PATTERNS IN DATA

```python
def f(x,y):
    return np.sin(x)**10 + np.cos(10+y+x)*np.cos(x)
```

##### Creating Z which is a function of both matrices

```python
Z = f(X,Y)
```

```python
Z.shape
```

```python
plt.contour(X,Y,Z, colors = 'Green')
plt.colorbar()
```

#### Here we are representing numbers in patterns and colors. Red means lesser values and black vice versa. Like valleys and hills!

```python
plt.contour(X,Y,Z,55, cmap='RdGy')
plt.colorbar()
plt.savefig('potli.jpg')
```
