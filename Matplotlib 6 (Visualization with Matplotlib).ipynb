# Visualization with Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,50)
```

```python
x
```

```python
plt.plot(x,np.sin(x),'o')
plt.plot(x,np.cos(x),'*')
```

```python
plt.subplot(1,3,1) #rows, col, panel-number
plt.plot(x,np.sin(x))
plt.subplot(1,3,2)
plt.plot(x,np.tan(x))
plt.subplot(1,3,3)
plt.plot(x,np.cos(x))
```
